You can use functions from the math module to help implement some of the equations above.

```python
x2 = range(0, 10)
y2 = [sqrt(n) for n in x2]
plt.plot(x2, y2)
plt.grid(c = 'slategray', linestyle = '-', linewidth = .5)
plt.title('y = sqrt(x)')
plt.axhline(y = 0, c = 'black')
plt.axvline(x = 0, c = 'black')
```

```python
x3 = range(-10, 10)
y3 = [n ** 3 for n in x3]

plt.plot(x3, y3)
plt.grid(c = 'slategray', linestyle = '-', linewidth = .5)
plt.title('y = x^3')
plt.axhline(y = 0, c = 'black')
plt.axvline(x = 0, c = 'black')
```

```python
x4 = range(-10, 10)
y4 = [2 ** n for n in x4]

plt.plot(x4, y4)
plt.grid(c = 'slategray', linestyle = '-', linewidth = .5)
plt.title('y = 2^x')
plt.axhline(y = 0, c = 'black')
plt.axvline(c = 'black')
```

```python
n_rows = 2
n_cols = 2

plt.subplot(n_rows, n_cols, 1)
plt.plot(x1, y1)
plt.title('y = x^2 - x + 2')
plt.grid(c = 'slategray', linestyle = '-', linewidth = .5)
plt.axhline(y = 0, c = 'black')
plt.axvline(c = 'black')

plt.subplot(n_rows, n_cols, 2)
plt.plot(x2, y2)
plt.title('y = sqrt(x)')
plt.grid(c = 'slategray', linestyle = '-', linewidth = .5)
plt.axhline(y = 0, c = 'black')
plt.axvline(c = 'black')

plt.subplot(n_rows, n_cols, 3)
plt.plot(x3, y3)
plt.title('y = x^3')
plt.grid(c = 'slategray', linestyle = '-', linewidth = .5)
plt.axhline(y = 0, c = 'black')
plt.axvline(c = 'black')

plt.subplot(n_rows, n_cols, 4)
plt.plot(x4, y4)
plt.title('y = 2^x')
plt.grid(c = 'slategray', linestyle = '-', linewidth = .5)
plt.axhline(y = 0, c = 'black')
plt.axvline(c = 'black')

plt.tight_layout()
plt.show()
```
