```python
from math import *
import matplotlib.pyplot as plt
```

```python
x1 = range(-4, 10)
x2 = range(1, 10)
x3 = range(0,10)
y1 = [0 * n + 1 for n in x1]
y2 = [log(n) for n in x2]
y3 = [n for n in x1]
y4 = [n * log(n) for n in x2]
y5 = [n ** 2 for n in x1]
y6 = [2 ** n for n in x1]
y7 = [factorial(n) for n in x3]
y8 = [n ** n for n in x1]
plt.plot(x1, y1, label = '$O(1)$')
plt.plot(x2, y2, label = '$O(log n)$')
plt.plot(x1, y3, label = '$O(n)$')
plt.plot(x2, y4, label = '$O(n log n)$')
plt.plot(x1, y5, label = '$O(n^2)$')
plt.plot(x1, y6, label = '$O(2^n)$')
plt.plot(x3, y7, label = '$O(n!)$')
plt.plot(x1, y8, label = '$O(n^n)$')
plt.title("Big O Notation")
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.grid(c = 'slategray', linestyle = '-', linewidth = .5)
plt.axhline(y = 0, c = 'black')
plt.axvline(c = 'black')
plt.ylim(-10, 15)
plt.legend()
```
